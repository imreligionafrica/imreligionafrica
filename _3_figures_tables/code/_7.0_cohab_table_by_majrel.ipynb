```python
import numpy as np
import pandas as pd
import re
import os
```

```python
current_folder = globals()['_dh'][0]
rootdir = os.path.dirname(os.path.dirname(current_folder))
wdir = os.path.join(rootdir, '_2_intermediate', 'data')
outdir = os.path.join(rootdir, '_3_figures_tables', 'data')
```

```python
df = pd.read_csv(os.path.join(wdir,'_cohab_cbyr.csv'))
df = df.fillna(0)
df['all_1418'] = df.all_1418.astype('int')
df['all_1425'] = df.all_1425.astype('int')
del df['bd']
df = df.groupby(['iso', 'year', 'major_religion']).sum().reset_index(drop=False)
df = df[~((df.major_religion == 'No Religion') | (df.major_religion == 'Other') | (df.major_religion == 0))].reset_index(drop=True)
df.index = [df.iso, df.year, df.major_religion]
del df['major_religion'], df['iso'], df['year']
df = df.unstack()
df.columns = [x[0] + '_' + x[1] for x in df.columns]
df = df.reset_index(drop=False)

tot = pd.DataFrame(df.sum(axis=0)).T
tot['iso'] = 'total'
tot['year'] = -9999
df = pd.concat([df, tot], axis=0)

df['cohab_any_1418_Christian'] = df.cohab_any_1418_Christian / df.all_1418_Christian
df['cohab_old_1418_Christian'] = df.cohab_old_1418_Christian / df.all_1418_Christian
df['cohab_any_1425_Christian'] = df.cohab_any_1425_Christian / df.all_1425_Christian
df['cohab_old_1425_Christian'] = df.cohab_old_1425_Christian / df.all_1425_Christian
df['cohab_any_1418_Muslim'] = df.cohab_any_1418_Muslim / df.all_1418_Muslim
df['cohab_old_1418_Muslim'] = df.cohab_old_1418_Muslim / df.all_1418_Muslim
df['cohab_any_1425_Muslim'] = df.cohab_any_1425_Muslim / df.all_1425_Muslim
df['cohab_old_1425_Muslim'] = df.cohab_old_1425_Muslim / df.all_1425_Muslim
df['cohab_any_1418_Traditional'] = df.cohab_any_1418_Traditional / df.all_1418_Traditional
df['cohab_old_1418_Traditional'] = df.cohab_old_1418_Traditional / df.all_1418_Traditional
df['cohab_any_1425_Traditional'] = df.cohab_any_1425_Traditional / df.all_1425_Traditional
df['cohab_old_1425_Traditional'] = df.cohab_old_1425_Traditional / df.all_1425_Traditional

for col in df.columns[2:]:
    if 'all' not in col:
        df[col] = df[col].astype('float')
        df[col] = np.round(100*df[col],3)
df = df[['iso',
         'year',   
         'all_1418_Christian',
         'cohab_any_1418_Christian',
         'cohab_old_1418_Christian',
         'all_1418_Muslim',
         'cohab_any_1418_Muslim',
         'cohab_old_1418_Muslim',
         'all_1418_Traditional',
         'cohab_any_1418_Traditional',
         'cohab_old_1418_Traditional',
         'all_1425_Christian',
         'cohab_any_1425_Christian',
         'cohab_old_1425_Christian',
         'all_1425_Muslim',
         'cohab_any_1425_Muslim',
         'cohab_old_1425_Muslim',
         'all_1425_Traditional',
         'cohab_any_1425_Traditional',
         'cohab_old_1425_Traditional']]
df.columns = ['iso', 'year'] + ['N', 'living with relative', 'living w/ old or fam mbr 15-40 yrs older']*6
df
```

# Writing the table

## Column names

```python
chlen = 16
incols = []
for col in df.columns:
    il = col.split(' ')
    # ilrev = il.copy()
    # ilrev.reverse()
    ol = []
    thislen = 0
    thisline = []
    for word in il:
        if len(word) + thislen <= chlen:
            thisline.append(word)
            thislen += len(word)
        else:
            ol.append(thisline)
            thislen = len(word)
            thisline = [word]
    ol.append(thisline)
    incols.append(ol)
maxlen = max([len(subl) for subl in incols])
for subl in incols:
    while len(subl) < maxlen:
        subl.append([])
outcols = []
for col in incols:
    outlines = []
    for line in col:
        outlines.append(" ".join(line))
    outcols.append(outlines)
```

```python
l1 = "\\begin{table}[ht!]\n"
l2 = "\singlespacing\n"
l3 = "\centering\n"
l4 = "\\caption{Cohabitation rates by census and major religion, ages 14-18 and 14-25}\n\\label{tab:cohab_1418_1425}\n\\resizebox{\columnwidth}{!}{\n"
l5 = "{|ll|"
for j in df.columns[2:]:
    l5+="{}".format('c')
l5 = '\\begin{tabular}' + l5 + "|}\n\\hline\n"
l6 = '&'
for i in range(len(df.columns[2:])):
               l6+='&({})'.format(i+1)
l6 += ' \\\\\n\\hline\n'
l7l = []
l7l.append([])
l7l.append([])
l7l.append([])
l7l.append([])
for idc, col in enumerate(outcols):
    for idl, line in enumerate(col):
        if idc == 0:
            l7l[idl].append(line)
        else:
            l7l[idl].append("&"+line)
            
l7 = ''
for l7i in l7l:
    l7+="".join(l7i)+"\\\\ \n"
l7+="\\hline\\hline \n"
```

```python
header = l1+l2+l3+l4+l5+l6+l7
```

## Body

```python
body = ''
rows = df.shape[0]
cols = df.shape[1]
for i in range(rows):
    row = ''
    for j in range(cols):
        if j == 0:
            row+="{}".format(df.iloc[i,j])
        elif j < 2:
            row+="&{}".format(df.iloc[i,j])
        elif 'N' in df.columns[j]:
            row+="&{:,}".format(df.iloc[i,j])
        else:
            if df.iloc[i,j] > 0:
                row += "&%s" % np.round(df.iloc[i,j], 1)              
            else:
                row+="&"
    row += "\\\\ \n\\hline\n"
    body +=row
body = body.replace('nan', '')
body = body.replace('.0&', '&')
```

## Footer

```python
l1 = "\\end{tabular}\n"
l2 = "}\n"
l3 = "\\captionsetup{justification=justified,width=\columnwidth,size=scriptsize}\n"
l4 = "\\caption*{"
l5 = '''DETAILED TABLE NOTES.'''
l6 = "}\n"
l7 = "\\end{table}"
```

```python
footer = l1+l2+l3+l4+l5+l6+l7
```

## Final cleanup and writing

```python
textab = header+body+footer
```

```python
textab = textab.replace('cccccccccccccccccc', 'ccc|ccc|ccc|ccc|ccc|ccc')
nl = '&&\\\\multicolumn{9}{c|}{\\\\textbf{Ages 14-18}}&\\\\multicolumn{9}{c|}{\\\\textbf{Ages 14-25}} \\\\\\ \n \\\\cline{3-11}\\\\cline{12-20}\n'
textab = re.sub(r'(\\hline\n)(&&)', r'\1%s\2' % nl, textab)
nl1 = '&&\\\\multicolumn{3}{c|}{\\\\textbf{Christian}}&\\\\multicolumn{3}{c|}{\\\\textbf{Muslim}}&\\\\multicolumn{3}{c|}{\\\\textbf{Traditional}}'
nl2 = '&\\\\multicolumn{3}{c|}{\\\\textbf{Christian}}&\\\\multicolumn{3}{c|}{\\\\textbf{Muslim}}&\\\\multicolumn{3}{c|}{\\\\textbf{Traditional}}'
nl3 = '\\\\\\ \n \\\\cline{3-11}\\\\cline{12-20}\n'
nl = nl1+nl2+nl3
textab = re.sub(r'(\\cline\{12-20\}\n)(&&)', r'\1%s\2' % nl, textab)
textab = re.sub('\\\\\\\\\s\n\\\\\\\\\s\n\\\\\\\\\s\n', '\\\\\\\\\n', textab)
textab = re.sub('-9999', '', textab)
```

```python
fh = open(outdir + "/_0_cohabtab_majrel_14_1825.tex", "w")
fh.write(textab)
fh.close()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.stats import norm, t
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
import country_converter as coco
import seaborn as sns
from CoefPlot import CoefPlot
import re
import os
```

```python
current_folder = globals()['_dh'][0]
rootdir = os.path.dirname(os.path.dirname(current_folder))
wdir = os.path.join(rootdir, '_2_intermediate', 'data')
outdir = os.path.join(rootdir, '_3_figures_tables', 'data')
```

```python
print(t.ppf(0.975, 1000))
print(norm.ppf(0.975))
```

```python
dfall = pd.read_csv(os.path.join(wdir, '_indlevel_balancedness_results.csv'))
dfall.loc[dfall.fe == 'cbd' , 'fe'] = 'country-birth-decade FEs'
dfall.loc[dfall.fe == 'cbd+dur' , 'fe'] = 'country-bd + district/urban-rural FEs'
dfall['fe'] = dfall.fe + ', ' + dfall.religion.str.title()



dfall['se'] = dfall['se'].astype(np.float64)
dfall['err_t'] = dfall.se*t.ppf(0.975, dfall.N)

df1 = dfall[(dfall['var'].isin(['nhhmemb', 'n_sameg_r', 'n_sameg_n', 'n_prevg_r', 'n_prevg_1540',
                                'n_prevg_r_nomompop', 'n_prevg_1540_nomompop'])) &
            (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df2 = dfall[(dfall['var'] == 'mg') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df3 = dfall[(dfall['var'].isin(['momonly', 'poponly', 'mompoponly', 'othersonly', 'extended'])) &
            (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df4 = dfall[(dfall['var'].isin(['rh_child', 'rh_biochild', 'rh_othchild', 'rh_grndchild',
                                'rh_head', 'rh_spouse', 'rh_sibl', 'rh_othrel', 'rh_nonrel'])) &
            (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df5 = dfall[(dfall['var'] == 'urban') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df6 = dfall[(dfall['var'] == 'ind0') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df7 = dfall[(dfall['var'] == 'occ0') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df8 = dfall[(dfall['var'] == 'a0_birth') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df9 = dfall[(dfall['var'] == 'amom_birth') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
df10 = dfall[(dfall['var'] == 'apop_birth') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
# df9 = dfall[(dfall['var'] == 'mig') & (dfall['sample'] == 'all')].reset_index(drop=True).copy(deep=True)
```

# 1) Basic household size 

```python
df1['variable'] = ''
df1.loc[df1['var'] == 'nhhmemb', 'variable'] = '# hh members'
df1.loc[df1['var'] == 'n_sameg_r', 'variable'] = '# hh members\nsame gen (rel head)'
df1.loc[df1['var'] == 'n_sameg_n', 'variable'] = '# hh members\nsame gen (age)'
df1.loc[df1['var'] == 'n_prevg_r', 'variable'] = '# hh members\nprevious gen (rel head)'
df1.loc[df1['var'] == 'n_prevg_1540', 'variable'] = '# hh members\nprevious gen (age)'
df1.loc[df1['var'] == 'n_prevg_r_nomompop', 'variable'] = '# hh members\nprevious gen (rel head)\nexcl mother/father'
df1.loc[df1['var'] == 'n_prevg_1540_nomompop', 'variable'] = '# hh members\nprevious gen (age)\nexcl mother/father'
# within_group_order = list(df1.variable[:7].values)
df1['order'] = list(range(7)) * 4
```

```python
df1
```

```python
cp = CoefPlot(df1)
f = cp.grouped_dotplot('b', 'fe', 'variable', err='err_t', figsize=(10,6), 
                        order_groups_by_variable='order',
                        within_group_order=['country-birth-decade FEs, Muslim',
                                            'country-birth-decade FEs, Traditional',
                                            'country-bd + district/urban-rural FEs, Muslim',
                                            'country-bd + district/urban-rural FEs, Traditional'],
                        zero_line=True, group_separators=True,
                        legend=True, legend_bbox=(0,-0.4), legend_columns=1)
f.savefig('{}/_4f1_balancedness_hhsz.pdf'.format(outdir), bbox_inches='tight', dpi=100)
plt.close(f)
#plt.show()
```
